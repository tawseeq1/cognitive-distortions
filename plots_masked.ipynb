```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import timedelta
def moving_average(data, window_size):
  moving_averages = []
  for i in range(window_size - 1, len(data)):
    window_sum = sum(data[i - window_size + 1: i + 1])
    moving_average = window_sum / window_size
    moving_averages.append(moving_average)
  return moving_averages
def filter_and_identify_spikes(data, window_size=4):
    filtered_data = []
    spikes = []
    for i in range(len(data)):
        if i >= window_size:
            avg_prev = np.mean(data[i - window_size:i])
            std_dev = np.std(data[i - window_size:i])
            if data[i] - avg_prev > std_dev:  # Adjusted condition: SD = 1
                spikes.append(i)
                filtered_data.append(data[i])
            else:
                filtered_data.append(0)  # Set non-spike values to 0
        else:
            filtered_data.append(0)  # Set initial values to 0
    return np.array(filtered_data), spikes
```
