```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
all_comments = pd.read_csv('/content/drive/MyDrive/Cognitive_Distortions_DrFarhanAli_Tawseeq/Teenagers/EXCEL/teen_all_comments.csv')
all_posts = pd.read_csv('/content/drive/MyDrive/Cognitive_Distortions_DrFarhanAli_Tawseeq/Teenagers/EXCEL/teen_all_posts.csv')
before_comments = pd.read_csv('/content/drive/MyDrive/Cognitive_Distortions_DrFarhanAli_Tawseeq/Teenagers/EXCEL/teen_comments_after.csv')
before_posts = pd.read_csv('/content/drive/MyDrive/Cognitive_Distortions_DrFarhanAli_Tawseeq/Teenagers/EXCEL/teen_posts_after.csv')
during_comments = pd.read_csv('/content/drive/MyDrive/Cognitive_Distortions_DrFarhanAli_Tawseeq/Teenagers/EXCEL/teen_comments_during.csv')
during_posts = pd.read_csv('/content/drive/MyDrive/Cognitive_Distortions_DrFarhanAli_Tawseeq/Teenagers/EXCEL/teen_posts_during.csv')
after_comments = pd.read_csv('/content/drive/MyDrive/Cognitive_Distortions_DrFarhanAli_Tawseeq/Teenagers/EXCEL/teen_comments_before.csv')
after_posts = pd.read_csv('/content/drive/MyDrive/Cognitive_Distortions_DrFarhanAli_Tawseeq/Teenagers/EXCEL/teen_posts_before.csv')
```

```python
merged_before = pd.concat([before_comments, before_posts], ignore_index=True)
merged_during = pd.concat([during_comments, during_posts], ignore_index=True)
merged_after = pd.concat([after_comments, after_posts], ignore_index=True)
merged_all = pd.concat([all_comments, all_posts], ignore_index=True)
```

```python
corr_merged_before = merged_before.iloc[1:, 1:-1].corr()
corr_merged_during = merged_during.iloc[1:, 1:-1].corr()
corr_merged_after = merged_after.iloc[1:, 1:-1].corr()
corr_merged_all = merged_all.iloc[1:, 1:-1].corr()
```

```python
corr_merged_before.to_csv('/content/drive/MyDrive/Cognitive_Distortions_DrFarhanAli_Tawseeq/Results/teenagers/teen_before_corr_percomment.csv')
corr_merged_during.to_csv('/content/drive/MyDrive/Cognitive_Distortions_DrFarhanAli_Tawseeq/Results/teenagers/teen_during_corr_percomment.csv')
corr_merged_after.to_csv('/content/drive/MyDrive/Cognitive_Distortions_DrFarhanAli_Tawseeq/Results/teenagers/teen_after_corr_percomment.csv')
corr_merged_all.to_csv('/content/drive/MyDrive/Cognitive_Distortions_DrFarhanAli_Tawseeq/Results/teenagers/teen_all_corr_percomment.csv')
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_merged_before, annot=True, fmt=".2f", cmap='coolwarm', vmin=-0.01, vmax=1)
plt.title('Correlation Matrix Before COVID-19 per comment _ Teenagers')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_merged_during, annot=True, fmt=".2f", cmap='coolwarm', vmin=-0.01, vmax=1)
plt.title('Correlation Matrix During COVID-19 per comment _ Teenagers')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_merged_after, annot=True, fmt=".2f", cmap='coolwarm', vmin=-0.01, vmax=1)
plt.title('Correlation Matrix After COVID-19 per comment _ Teenagers')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_merged_all, annot=True, fmt=".2f", cmap='coolwarm', vmin=-0.01, vmax=1)
plt.title('Correlation Matrix All per comment _ Teenagers')
plt.show()
```
